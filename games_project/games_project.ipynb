# Исследование продаж компьютерых игр

Интернет-магазин «Стримчик» продаёт по всему миру компьютерные игры. Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation). Нужно выявить определяющие успешность игры закономерности. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.

**Цель исследования:** выявить определяющие успешность игры закономерности, чтобы сделать ставку на потенциально популярный продукт и спланировать рекламные кампании на следующий год.

**Исходные данные:** исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы.

**Описание исходных данных:** \
`Name` — название игры \
`Platform` — платформа \
`Year_of_Release` — год выпуска \
`Genre` — жанр игры \
`NA_sales` — продажи в Северной Америке (миллионы проданных копий) \
`EU_sales` — продажи в Европе (миллионы проданных копий) \
`JP_sales` — продажи в Японии (миллионы проданных копий) \
`Other_sales` — продажи в других странах (миллионы проданных копий) \
`Critic_Score` — оценка критиков (максимум 100) \
`User_Score` — оценка пользователей (максимум 10) \
`Rating` — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию. 

**План исследования:**
1. Общий обзор исходных данных. 
2. Предобработка данных.
3. Исследовательский анализ данных.
4. Исследование пользователей из каждого региона продаж.
5. Проверка гипотез.
6. Общий вывод.

*Из столбца `name` видно, что в некоторые названия игр входят числовые значения, похожие на годы. По-видимому, эти значения означают год выпуска данной версии игры. Если извлечь из названия год, можно заполнить пропуск в столбце `year_of_release` в этой строке.*

*Выведем статистику по столбцу `year_of_release`, чтобы узнать диапазон лет.*

```python
data['year_of_release'].describe()
```

*Создадим функцию `fill_years_of_release`, которая с помощью регулярных выражений находит в строке столбца `name` четырёхзначное число, сравнивает его с годами из диапазона, извлекает и возвращает это значение.*

```python
def fill_years_of_release(name):
    if re.search(r'\d{4}', name) and \
    1980 <= int(name[re.search(r'\d{4}', name).start():re.search(r'\d{4}', name).end()]) <= 2016:
        return int(name[re.search(r'\d{4}', name).start():re.search(r'\d{4}', name).end()])
```

*Применим функцию `fill_years_of_release` к тем строкам, где имеются пустые значения в столбце `year_of_release`.*

```python
data['year_of_release'].loc[data['year_of_release'].isna()] = \
                     data.loc[data['year_of_release'].isna()]['name'].apply(fill_years_of_release)
```

## Проверка гипотез

### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

*Сформулируем нулевую и альтернативную гипотезы. Мы предполагаем, средние оценки пользователей у платформ Xbox One и PC не отличаются, и принимаем это утверждение за нулевую гипотезу. Соответственно, за альтернативную гипотезу мы принимаем утверждение, что средние оценки пользователей у платформ Xbox One и PC разные.*

H0: средние пользовательские рейтинги платформ Xbox One и PC одинаковые; \
H1: средние пользовательские рейтинги платформ Xbox One и PC разные.    

*Будем использовать метод для проверки гипотезы о равенстве среднего двух генеральных совокупностей по взятым из них выборкам.*

*Сформируем массив с 1-й выборкой.*

```python
xone_user_score = data.query('platform == "XOne" and user_score != -1')['user_score']
```

*Сформируем массив со 2-й выборкой.*

```python
pc_user_score = data.query('platform == "PC" and user_score != -1')['user_score']
```

*Определим размеры выборок.*

```python
len(xone_user_score)
```

```python
len(pc_user_score)
```

*Размеры выборок отличаются, поэтому примем значение параметра equal_var в используемом методе равным False.*

*Задаём критический уровень значимости.*

```python
alpha = 0.05
```

*Проверяем гипотезы.*

```python
from scipy import stats as st
results = st.ttest_ind(
    xone_user_score,
    pc_user_score,\
    equal_var = False)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

### Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные

*Сформулируем нулевую и альтернативную гипотезы. За нулевую гипотезу примем предположение, что средние оценки пользователей у жанров Action и Sports не отличаются. Соответственно, за альтернативную гипотезу мы принимаем утверждение, что средние оценки пользователей у жанров Action и Sports разные.*

H0: средние пользовательские рейтинги жанров Action и Sports одинаковые; \
H1: средние пользовательские рейтинги жанров Action и Sports разные.  

*Будем использовать метод для проверки гипотезы о равенстве среднего двух генеральных совокупностей по взятым из них выборкам.*

*Сформируем массив с 1-й выборкой.*

```python
action_user_score = data.query('genre == "Action" and user_score != -1')['user_score']
```

*Сформируем массив со 2-й выборкой.*

```python
sports_user_score = data.query('genre == "Sports" and user_score != -1')['user_score']
```

*Определим размеры выборок.*

```python
len(action_user_score)
```

```python
len(sports_user_score)
```

*Размеры выборок разные, поэтому примем значение параметра equal_var в используемом методе равным False.*

*Задаём критический уровень значимости.*

```python
alpha = 0.05
```

*Проверяем гипотезы.*

```python
results = st.ttest_ind(
    action_user_score, 
    sports_user_score,
    equal_var = False)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

**Отвергаем нулевую гипотезу, а это значит, что средние пользовательские рейтинги жанров Action и Sports разные.**
