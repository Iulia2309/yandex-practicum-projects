# Исследование рынка недвижимости Санкт-Петербурга

На основе этих данных можно определять рыночную стоимость объектов недвижимости. Задача работы — установить параметры. Это позволит построить автоматизированную систему: она отследит аномалии и мошенническую деятельность. 

По каждой квартире на продажу доступны два вида данных. Первые вписаны пользователем, вторые — получены автоматически на основе картографических данных. Например, расстояние до центра, аэропорта, ближайшего парка и водоёма. 

**Цель исследования:** установить параметры, влияющие на рыночную стоимость объектов недвижимости.

**Задачи исследования:**

1. Изучить параметры объектов недвижимости. 
2. Изучить скорость продажи объектов недвижимости.

5. Изучить, как меняется общая стоимость объектов недвижимости по мере удаления от цента Санкт-Петербурга. 

**Исходные данные** — архив объявлений о продаже квартир в Санкт-Петербурге и соседних населённых пунктах за несколько лет.

**Заполним пропуски, где это возможно.**

*Данные в столбцах `parks_nearest` и `ponds_nearest` должны были определяться автоматически по картографическим данным.
Возможно, пропуски в этих столбцах означают, что парков и водоёмов в шаговой доступности нет.*

*Выведем информацию по срезу, в котором значения в столбце `parks_around3000` равны 0, т.е. отсутствуют парки в радиусе 3 км.*

```python
data.query('parks_around3000 == 0').info()
```

*Как видно, в столбце `parks_nearest` для случаев, когда значения в столбце `parks_around3000` равны 0, есть только 4
ненулевых значения. Выведем эти стоки:*

```python
data.loc[(data['parks_around3000'] == 0) & (~data['parks_nearest'].isna())]
```

*В этих четырёх строках расстояние до ближайшего парка оказалось чуть больше 3000 м. Следует отметить, что столбец `parks_around3000` содержит 5518 пустых значений,а значит, для таких записей мы не можем однозначно сказать, есть в шаговой доступности парк или нет. Поскольку для исследования нам нужно только расстояние до парка, расположенного в шаговой доступности, нет смысла заменять все пропуски в столбце `parks_nearest`. Оставим их как есть.*

*Нам больше не понадобится столбец `parks_around3000`, поэтому удалим его.*

```python
data = data.drop('parks_around3000', axis=1)
```

*Пропуски в столбце `balcony`, очевидно, означают, что балконы у данного объекта недвижимости отсутствуют. \
Заменим пропуски на значения 0.*

```python
data['balcony'] = data['balcony'].fillna(0)
```

*Пропуски в столбце `ceiling_height` заменим медианой, поскольку она более устойчива к изменению минимальных и максимальных значений, чем среднее. Поскольку нам неизвестны типы домов и годы постройки, медиану придётся взять по всем значениям столбца.*

```python
data.loc[data['ceiling_height'].isna(), 'ceiling_height'] = data['ceiling_height'].median()
```

*Значения в столбцах `airports_nearest` и `cityCenters_nearest` должны были определяться автоматически по картографическим данным. Пропусков в этих столбцах примерно одинаковое количество и 
наблюдаются они для одних и тех же объектов. Такие же пропуски были и в удалённых столбцах `parks_around300` и `ponds_around300`. Возможно, алгоритм определения расстояний по карте не сработал, или
произошла ошибка при подготовке данных. В реальном исследовании этот был бы повод для составления баг-репорта.*

*Можно было бы для небольших населённых пунктов заполнить пропущенные значения в столбцах `airports_nearest` и `cityCenters_nearest` медианными значениями по имеющимся данным для каждого населённого пункта. Для Санкт-Петербурга этот вариант не подходит вследствие значительной площади его территории. Но, поскольку исследование взаимосвязи стоимости и расстояния от центра мы будем проводить только для объектов недвижимости Санкт-Петербурга, смысла делать такие замены нет. Оставим пропуски.*

*Пропуски в столбце `days_exposition` могут свидетельствовать о техническом сбое или ошибке при подготовке данных. \
Пропуски заменим медианой по всем значениям столбца, поскольку она более устойчива к изменению минимальных и максимальных значений, чем среднее.*

```python
data.loc[data['days_exposition'].isna(), 'days_exposition'] = data['days_exposition'].median()
```

*Пропуски в столбцax `kitchen_area` и `living_area` могли появиться в результате незаполнения пользователем соответствующих пунктов.*

*Мы можем принять, что в квартирах-студиях площадь кухни равна 0.*

```python
data.loc[(data['studio']) & (data['kitchen_area'].isna()), 'kitchen_area'] = 0
```

*В столбцах с общей площадью `total_area` и количеством комнат `rooms` пропусков нет, а значит, эти столбцы можно использовать для заполнения пропусков в столбцах `kitchen_area` и `living_area`. По данным строк, для которых известны значения всех трёх переменных - `total_area`, `kitchen_area` и `living_area` , найдём соотношения living_area/total_area и kitchen_area/total_area для разного количества комнат. Создадим отдельный датафрейм, в которрый включим только те строки, где общая площадь больше или равна сумме площади кухни и жилой площади.*

```python
data_area = data.query('~living_area.isna() and ~kitchen_area.isna() and total_area >= living_area + kitchen_area',
                       engine='python')[['rooms', 'total_area', 'kitchen_area', 'living_area']]
```

*Добавим в `data_area` столбцы с отношениями living_area/total_area и kitchen_area/total_area для разного количества комнат.*

```python
data_area['living_to_total'] = data_area['living_area'] / data_area['total_area']
```

```python
data_area['kitchen_to_total'] = data_area['kitchen_area'] / data_area['total_area']
```

*Создадим сводную таблицу по данным `data_area`, в которой посчитаем средние значения отношений living_area/total_area и kitchen_area/total_area для разного количества комнат*

```python
data_area.pivot_table(index='rooms', values=['kitchen_to_total', 'living_to_total'], aggfunc='mean')
```

*Посмотрим, что может означать количество комнат, равное 0. Выведем 10 таких строк.*

```python
data.query('rooms == 0').head(10)
```

*Очевидно, что значения `rooms`, равные нулю, соответствуют либо квартире-студии, либо квартире со свободной планировкой.*

*Выведем уникальные значения количества комнат для случаев пропусков в столбцах.*

```python
data.query('kitchen_area.isna()', engine='python')['rooms'].unique()
```

```python
data.query('living_area.isna()', engine='python')['rooms'].unique()
```

*Пропуски в столбцах `kitchen_area` и `living_area` наблюдаются для количества комнат не более 9. \
Создадим два списка, в которые запишем соотношения living_area/total_area и kitchen_area/total_area для количества комнат из pivot_table от 0 до 9. Индексы списков соответствуют количеству комнат.*

```python
kitchen_to_total = list(data_area.pivot_table(index='rooms', \
                                              values=['living_to_total', 'kitchen_to_total'], 
                                              aggfunc='mean') \
                                 .values[:10, 0])
```

```python
living_to_total = list(data_area.pivot_table(index='rooms', \
                                             values=['living_to_total', 'kitchen_to_total'],
                                             aggfunc='mean') \
                                .values[:10, 1])
```

*Выведем созданные списки и проверим правильность их заполнения.*

```python
kitchen_to_total
```

```python
living_to_total
```

*Заполним пустые значения в столбце `kitchen_area`, умножив значение `total_area` на соответствующий коэффициент из списка kitchen_to_total.*

```python
data['kitchen_area'].loc[data['kitchen_area'].isna()] = \
    data.loc[data['kitchen_area'].isna()].apply(lambda x: round(kitchen_to_total[x['rooms']] * x['total_area'], 1), axis=1)
```

*Заполним пустые значения в столбце `living_area`, умножив значение `total_area` на соответствующий коэффициент из списка living_to_total.*

```python
data['living_area'].loc[data['living_area'].isna()] = \
    data.loc[data['living_area'].isna()].apply(lambda x: round(living_to_total[x['rooms']] * x['total_area'], 1), axis=1)
```

*Пропуски в столбце `floors_total` и `locality_name` могли появиться вследствие незаполнения поля пользователем.*

*Пропуски в столбце `floors_total` было бы некорректно заменять медианой по всем значениям столбца, т.к. могли бы возникнуть строки, где общее количество этажей меньше этажа, на котором расположена квартира. Попробуем заменить пропуски медианой по уникальным значениям столбца `floor`.*

```python
for f in data['floor'].unique():
    data.loc[(data['floor'] == f) & (data['floors_total'].isna()), 'floors_total'] = \
    data.loc[(data['floor'] == f), 'floors_total'].median()
```

### Вычисление средней стоимости объекта недвижимости в зависимости от расстояния от центра Санкт-Петербурга

*Выведем среднюю стоимость квартир для каждого километра расстояния от центра города. Поскольку нас интересует стоимость квартир только в пределах Санкт-Петербурга без учёта пригородов, ограничим расстояние от центра города 18 километрами.*

```python
data.query('locality_name == "Санкт-Петербург" and cityCenters_nearest_km <= 18') \
    .pivot_table(index='cityCenters_nearest_km', 
                 values='last_price_mln', 
                 aggfunc='median') \
    .round(2)
```

*Построим график изменения стоимости объектов недвижимости в зависимости от расстояния от цента города.*

```python
data.query('locality_name == "Санкт-Петербург" and cityCenters_nearest_km <= 18') \
    .groupby('cityCenters_nearest_km')['last_price_mln'] \
    .median() \
    .plot(y='price', 
          style='o-',
          linewidth=3, 
          figsize=(10, 8),
          grid=True) 
plt.xlabel('Расстояние от центра, км', fontsize=14)
plt.ylabel('Стоимость, млн', fontsize=14)
plt.title('Изменение стоимости в зависимости от расстояния от цента СПб', fontsize=16)
plt.show()
```

**Стоимость объектов недвижимости в целом имеет тенденцию к уменьшению при увеличении расстояния от цента Санкт-Петербурга. При этом в самом центре города (расстояние до 1 км) средняя стоимость объектов недвижимости увеличивается, достигая пика на отметке 1 км. После этого до отметки в 3 км отмечается резкое снижение средней стоимости, затем до отметки в 5 км сохраняются примерно одинаковые значения, после чего снова резкое снижение до отметки 9 км. Затем стоимость стабилизируется и сохнаряет примерно одинаковые значения до отметки 15 км, после чего плавно уменьшается.**
